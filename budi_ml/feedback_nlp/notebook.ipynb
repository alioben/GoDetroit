```python
import pandas as pd
```

```python
park = pd.read_csv('Parks_2016.csv')
```

```python
park['the_geom'] = park['the_geom'].apply(lambda x: x.split(' ')[1:])
```

```python
park['the_geom'] = park['the_geom'].apply(lambda x: x[0:2])
```

```python
park['LAT'] = park['the_geom'].apply(lambda x: x[1])
```

```python
park['LON'] = park['the_geom'].apply(lambda x: x[0])
```

```python
park['LON'] = park['LON'].apply(lambda x:x[3:])
```

```python
park['LON'] = park['LON'].astype('float64')
```

```python
park['LAT'] = park['LAT'].apply(lambda x:x[:-1]).astype('float64')
```

```python
park = park[['LAT', 'LON']]
```

```python
import numpy as np
```

```python
pd.options.mode.chained_assignment = None 
park['SCORE'] = pd.Series(np.random.normal(50000, 35)).copy()
```

```python
park = park.sample(frac=1)
park['SCORE'] = 0
```

```python
park_1 = park.head(50)
lat_array = np.array(park_1['LAT'])
lon_array = np.array(park_1['LON'])
```

```python
combined = []
```

```python
for _ in range(len(lat_array)):
    combined.append([lat_array[_], lon_array[_]])
```

```python
combined = np.array(combined)
np.savetxt(fname='list_of_parks8.txt', X=combined, delimiter=',', fmt='%5.7f')
```
